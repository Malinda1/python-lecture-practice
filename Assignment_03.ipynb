inventory management system for a small sotor

```python

# 01. Sample inventory initialization
inventory = [
    {'name': 'Apple', 'quantity': 50, 'price': 450},
    {'name': 'Banana', 'quantity': 30, 'price': 560},
    {'name': 'Orange', 'quantity': 20, 'price': 350},
    {'name': 'Mango', 'quantity': 45, 'price': 255},
    {'name': 'iphone', 'quantity': 24, 'price': 75000},
    {'name': 'Macbook', 'quantity': 10, 'price': 15000},
    {'name': 'HP laptop', 'quantity': 12, 'price': 23000},
    {'name': 'redmi phone', 'quantity': 34, 'price': 456000}
]

# 02. Function to display the main menu
def display_menu():
    print("\nInventory Management System")
    print("1. Add a new item you want")
    print("2. Remove an item if you want")
    print("3. Update an item if you want")
    print("4. Display all products")
    print("5. Exit the program")

# 03. Function to add a new item
def add_item():
    name = input("Enter the name of the item: ")
    quantity = int(input("Enter the quantity of item: "))
    price = float(input("Enter the price of item: "))
    inventory.append({'name': name, 'quantity': quantity, 'price': price})
    print(f"{name} has been added to the inventory.")

# 04. Function to remove an item
def remove_item():
    name = input("Enter the name of the item to remove: ")
    for item in inventory:
        if item['name'].lower() == name.lower():
            inventory.remove(item)
            print(f"{name} has been removed from the inventory.")
            return
    print(f"Item '{name}' not found in the inventory.")

# 05. Function to update an item
def update_item():
    name = input("Enter the name of the item to update: ")
    for item in inventory:
        if item['name'].lower() == name.lower():
            new_quantity = int(input(f"Enter the new quantity for {name}: "))
            new_price = float(input(f"Enter the new price for {name}: "))
            item['quantity'] = new_quantity
            item['price'] = new_price
            print(f"{name} has been updated.")
            return
    print(f"Item '{name}' not found in the inventory.")

# 06. Function to display all items
def display_items():
    if not inventory:
        print("No items in the inventory.")
    else:
        print(f"{'Item Name':<15} {'Quantity':<10} {'Price':<10}")
        print("-" * 35)
        for item in inventory:
            print(f"{item['name']:<15} {item['quantity']:<10} {item['price']:<10.2f}")

# 07. Main program loop
def main():
    while True:
        display_menu()
        choice = input("Select an option (1-5): ")

        if choice == '1':
            add_item()
        elif choice == '2':
            remove_item()
        elif choice == '3':
            update_item()
        elif choice == '4':
            display_items()
        elif choice == '5':
            print("Exiting the application. Goodbye!")
            break
        else:
            print("Invalid option. Please try again.")

if __name__ == "__main__":
    main()
```
