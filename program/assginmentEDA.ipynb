```python
# Rename columns for easier reference
df = data01.rename(columns={
    'Ladder score': 'Happiness_Score',
    'Logged GDP per capita': 'GDP_per_capita',
    'Social support': 'Social_Support',
    'Healthy life expectancy': 'Life_Expectancy',
    'Freedom to make life choices': 'Freedom',
    'Perceptions of corruption': 'Corruption',
    'Standard error of ladder score': 'steols'
})
```

```python
# Convert data type into Integer 
df['Happiness_Score'] = df['Happiness_Score'].astype(int)
```

```python
# Drop irrelevant columns

df = df.drop(['Generosity', 'Explained by: Generosity'], axis = 1)
```

# What factors seem to influence happiness the most?

These countries rank the highest due to a combination of several key factors:

	1.	Logged GDP per capita: Economic prosperity is a strong predictor of happiness, as it ensures a higher standard of living.
	2.	Social support: All these countries score highly in terms of social support, which indicates that people feel they can rely on others during difficult times.
	3.	Healthy life expectancy: Good healthcare systems and higher life expectancy also contribute to happiness.
	4.	Freedom to make life choices: The freedom to make important life decisions (like employment, family, etc.) also plays a crucial role.
	5.	Low corruption perceptions: These countries have lower perceptions of corruption, which fosters trust in government and public institutions.


# Are there any notable trends or patterns?

1. Economic Prosperity (Logged GDP per Capita) and Happiness:

	•	Trend: Countries with higher GDP per capita tend to have higher happiness scores.
	•	Pattern: Many of the top-ranked countries, such as Finland, Denmark, Switzerland, and the Netherlands, all have strong economies with high GDP per capita. Economic stability provides better access to resources, services, and opportunities, contributing significantly to overall well-being.

2. Social Support is a Strong Predictor of Happiness:

	•	Trend: Social support is highly correlated with happiness. Countries where individuals feel they have someone to count on during tough times report higher happiness scores.
	•	Pattern: Top-ranking countries like Iceland and Finland score very high in social support, indicating that strong community and family ties are key contributors to happiness.

3. Health and Life Expectancy:

	•	Trend: Higher life expectancy, particularly healthy life expectancy, is closely tied to higher happiness levels.
	•	Pattern: Healthier populations with good healthcare systems, like in Switzerland and the Nordic countries, tend to have better overall happiness scores.
