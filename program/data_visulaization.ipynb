```python
import pandas as pd


# Group by 'Base Flavor' and calculate the mean for 'Flavor Rating', 'Texture Rating', and 'Total Rating'
grouped_by_base_flavor01 = data01.groupby('Base Flavor').agg(
    avg_flavor_rating=('Flavor Rating', 'mean'),
    avg_texture_rating=('Texture Rating', 'mean'),
    avg_total_rating=('Total Rating', 'mean'),
    sum_flavor_rating=('Flavor Rating', 'sum'),
    sum_texture_rating=('Texture Rating', 'sum'),
    sum_total_rating=('Total Rating', 'sum'),
    flavor_count=('Flavor', 'count'),
    max_flavor_rating=('Flavor Rating', 'max'),
    min_flavor_rating=('Flavor Rating', 'min'),
    max_texture_rating=('Texture Rating', 'max'),
    min_texture_rating=('Texture Rating', 'min')
).reset_index()

# Display the grouped data
grouped_by_base_flavor01
```

```python
final_grouped = grouped_by_base_flavor01.groupby('avg_flavor_rating').agg(
    sum_avg_flavor_rating=('avg_flavor_rating', 'sum'),
    mean_avg_flavor_rating=('avg_flavor_rating', 'mean'),
    min_avg_flavor_rating=('avg_flavor_rating', 'min'),
    max_avg_flavor_rating=('avg_flavor_rating', 'max'),
    sum_texture_rating=('avg_texture_rating', 'sum'),
    mean_texture_rating=('avg_texture_rating', 'mean'),
    min_texture_rating=('avg_texture_rating', 'min'),
    max_texture_rating=('avg_texture_rating', 'max'),
    sum_total_rating=('avg_total_rating', 'sum'),
    mean_total_rating=('avg_total_rating', 'mean'),
    min_total_rating=('avg_total_rating', 'min'),
    max_total_rating=('avg_total_rating', 'max')
).reset_index()

# Display the final grouped result
final_grouped
```

```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
import numpy as np
```

```python
data02 = pd.read_csv(r"/Users/pasindumalinda/Documents/python project/lecture practices/program/dataset/Ice Cream Ratings.csv")
```

```python
data02.head()
```

```python
data02.shape
```

```python
data02.info()
```

```python
data02['Date'] = pd.to_datetime(data02['Date'], errors='coerce')
```

```python
data02.info()
```

```python
plt.figure(figsize=(12, 6))  # Set the figure size

# Plot each rating against the Date
plt.plot(data02['Date'], data02['Flavor Rating'], marker='*', label='Flavor Rating', color='blue')
plt.plot(data02['Date'], data02['Texture Rating'], marker='*', label='Texture Rating', color='green')
plt.plot(data02['Date'], data02['Overall Rating'], marker='*', label='Overall Rating', color='orange')

# Adding titles and labels
plt.title('Ratings Over Time')
plt.xlabel('Date')
plt.ylabel('Rating')
plt.xticks(rotation=45)  # Rotate date labels for better visibility
plt.legend()  # Show legend
plt.grid()  # Show grid

# Display the plot
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()
```

```python
# 1. Flavor Rating
plt.figure(figsize=(10, 5))  # Set figure size
plt.plot(data02['Date'], data02['Flavor Rating'], marker='o', color='blue', label='Flavor Rating')
plt.title('Flavor Rating Over Time')
plt.xlabel('Date')
plt.ylabel('Flavor Rating')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# 2. Texture Rating
plt.figure(figsize=(10, 5))  # Set figure size
plt.plot(data02['Date'], data02['Texture Rating'], marker='o', color='green', label='Texture Rating')
plt.title('Texture Rating Over Time')
plt.xlabel('Date')
plt.ylabel('Texture Rating')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# 3. Overall Rating
plt.figure(figsize=(10, 5))  # Set figure size
plt.plot(data02['Date'], data02['Overall Rating'], marker='o', color='orange', label='Overall Rating')
plt.title('Overall Rating Over Time')
plt.xlabel('Date')
plt.ylabel('Overall Rating')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
data02.set_index('Date', inplace=True)
```

```python
data02.head()
```

```python
# Plotting the bar chart
bar_width = 0.25  # Width of the bars
x = np.arange(len(data02.index))  # The label locations

# Creating the bar chart
plt.figure(figsize=(12, 6))  # Set the figure size
plt.bar(x - bar_width, data02['Flavor Rating'], width=bar_width, label='Flavor Rating', color='blue')
plt.bar(x, data02['Texture Rating'], width=bar_width, label='Texture Rating', color='green')
plt.bar(x + bar_width, data02['Overall Rating'], width=bar_width, label='Overall Rating', color='orange')

# Adding titles and labels
plt.title('Ratings Over Time')
plt.xlabel('Date')
plt.ylabel('Rating')
plt.xticks(x, data02.index.date, rotation=45)  # Use the index for x-tick labels
plt.legend()  # Show legend
plt.grid(axis='y')  # Add gridlines for y-axis

# Display the plot
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()
```

```python
# Plotting the stacked bar chart
plt.figure(figsize=(12, 6))  # Set the figure size

# Creating the stacked bar chart
plt.bar(data02.index, data02['Flavor Rating'], label='Flavor Rating', color='blue')
plt.bar(data02.index, data02['Texture Rating'], bottom=data02['Flavor Rating'], label='Texture Rating', color='green')
plt.bar(data02.index, data02['Overall Rating'], bottom=data02['Flavor Rating'] + data02['Texture Rating'], label='Overall Rating', color='orange')

# Adding titles and labels
plt.title('Stacked Ratings Over Time')
plt.xlabel('Date')
plt.ylabel('Rating')
plt.xticks(rotation=45)  # Rotate x-tick labels for better visibility
plt.legend()  # Show legend
plt.grid(axis='y')  # Add gridlines for the y-axis

# Display the plot
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()
```

```python
# Plotting the scatter plot
plt.figure(figsize=(12, 6))  # Set the figure size

# Create scatter plots for each rating
plt.scatter(data02.index, data02['Flavor Rating'], label='Flavor Rating', color='blue', marker='o', s = 100)
plt.scatter(data02.index, data02['Texture Rating'], label='Texture Rating', color='green', marker='s', s = 100)
plt.scatter(data02.index, data02['Overall Rating'], label='Overall Rating', color='orange', marker='^', s = 100)

# Adding titles and labels
plt.title('Scatter Plot of Ratings Over Time')
plt.xlabel('Date')
plt.ylabel('Rating')
plt.xticks(rotation=45)  # Rotate x-tick labels for better visibility
plt.legend()  # Show legend
plt.grid()  # Add gridlines

# Display the plot
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()
```
