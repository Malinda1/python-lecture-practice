```python
import pandas as pd
```

```python
data01 = pd.read_csv(r"/Users/pasindumalinda/Documents/python project/lecture practices/program/dataset/Flavors.csv")
```

```python
data01
```

```python
data01.info()
```

```python
group_by_base_flavor = data01.groupby('Base Flavor').agg(
    avg_flavor_rating=('Flavor Rating', 'mean'),
    sum_flavor_rating=('Flavor Rating', 'sum'),
    flavor_count=('Flavor Rating', 'count'),
    max_flavor_rating=('Flavor Rating', 'max'),
    min_flavor_rating=('Flavor Rating', 'min'),
    avg_texture_rating=('Texture Rating', 'mean'),
    sum_texture_rating=('Texture Rating', 'sum'),
    texture_count=('Texture Rating', 'count'),
    max_texture_rating=('Texture Rating', 'max'),
    min_texture_rating=('Texture Rating', 'min'),
    sum_total_rating=('Total Rating', 'sum')
    
).reset_index()

group_by_base_flavor
```

```python
final_group = group_by_base_flavor.groupby('avg_flavor_rating').agg(
    sum_flavor_rating=('avg_flavor_rating', 'sum'),
    sum_texture_rating=('avg_texture_rating', 'sum'),
    min_flavor_rating=('avg_flavor_rating', 'min'),
    max_flavor_rating=('avg_flavor_rating', 'max'),
    min_texture_rating=('avg_texture_rating', 'min'),
    max_texture_rating=('avg_texture_rating', 'max'),
    mean_total_rating=('sum_total_rating', 'mean'),
    min_total_rating=('sum_total_rating', 'min'),
    max_total_rating=('sum_total_rating', 'max')
    
    
).reset_index()

final_group
```

```python
import matplotlib.pyplot as plt
```

```python
data02 = pd.read_csv(r"/Users/pasindumalinda/Documents/python project/lecture practices/program/dataset/Ice Cream Ratings.csv")
```

```python
data02
```

```python
data02.info()
```

```python
data02['Date'] = pd.to_datetime(data02['Date'], errors= 'coerce')
```

```python
#create a figure size
plt.figure(figsize= (12, 6))

#Drawing plot
plt.plot(data02['Date'], data02['Flavor Rating'], marker = 'o', label = 'Flavor Rating', color = 'blue')
plt.plot(data02['Date'], data02['Texture Rating'], marker = 'o', label = 'Texture Rating', color = 'green')
plt.plot(data02['Date'], data02['Overall Rating'], marker = 'o', label = 'Overall Rating', color = 'orange')

#adding title and label 
plt.title('Rating over Date')
plt.xlabel('Date')
plt.ylabel('Rating')
plt.xticks(rotation = 45)
plt.legend()
plt.grid()

#Show created plot
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize= (12, 6))

plt.plot(data02['Date'], data02['Flavor Rating'], marker = 's', label = 'Flavor Rating', color = 'blue')

plt.title('Flavor Rating over Date')
plt.xlabel('Date')
plt.ylabel('Flavor Rating')
plt.legend()
plt.xticks(rotation = 45)
plt.grid()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize= (12, 6))

plt.plot(data02['Date'], data02['Texture Rating'], marker = 's', label = 'Texture Rating', color = 'green')

plt.title('Texture Rating over Date', fontweight = 'bold', color = 'darkblue', fontsize = 18)
plt.xlabel('Date', fontweight = 'bold', color = '#387478', fontsize = 15)
plt.ylabel('Texture Rating', fontweight = 'bold', color = '#387478', fontsize = 15)
plt.xticks(rotation = 35)
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (12, 6))

plt.plot(data02['Date'], data02['Overall Rating'], marker = 's', label = 'Overall Rating', color = 'orange')

plt.title('Overall Rating over Date', fontweight = 'bold', color = '#243642', fontsize = 18)
plt.xlabel('Date', fontweight = 'bold', color = '#629584', fontsize = 15)
plt.ylabel('Overall Rating', fontweight = 'bold', color = '#629584', fontsize = 15)
plt.xticks(rotation = 35)
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
